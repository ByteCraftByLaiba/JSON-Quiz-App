## Function to write data from python to json file in 'RegisteredUser.json'

The **json.dump()** function is used to serialize Python objects to a JSON formatted string and write it to a file-like object. The name "dump" stands for "dump to file". It is part of the json module, which provides a convenient way to work with JSON data.Now, in this function we're going to write the registered users in 'RegisteredUser.json' using json.dump().

```python
# Function to write in "RegisteredUser.json" using dump()
def writeInRegisteredUserFile(UsersDict):
    with open("RegisteredUser.json", "w") as outfile:
        json.dump(UsersDict, outfile, indent=4)
```

## Function to read data from json file 'RegisteredUser.json' into python 

The **json.load()** function is used to deserialize a JSON file or a JSON string. The name "load" stands for "load from file." It is part of the json module, which provides a convenient way to work with JSON data.Now, in this function we're going to read the registered users from 'RegisteredUser.json' using json.load() to check whether the user trying to login is registered or not.

```python
# Function to read from "RegisteredUser.json" using load()
def readFromRegisteredUserFile(UsersList):
    with open("RegisteredUser.json", "r") as registeredUsers:
        UsersDict = json.load(registeredUsers)
    UsersList = UsersDict["User"]
    return UsersList
```

## Function to read data from json file 'QuestionAnswer.json' into python 

The **json.load()** function is used to deserialize a JSON file or a JSON string. The name "load" stands for "load from file." It is part of the json module, which provides a convenient way to work with JSON data.Now, in this function we're going to read the Multiple Choice Questions and their coreect answer from 'QuestionAnswer.json' using json.load().

```python
# Function to read from "QuestionAnswer.json" using load()
def readFromQuestionAnswerFile():
    with open("QuestionAnswer.json", "r") as QA:
        questionsDict = json.load(QA)
    return questionsDict
```

## Main Menu of Quiz App
Now, we're going to make a quiz application. First of all, user will sign up to take the quiz and store his/her information in the file named **registeredUser.json**. If already registered, then this program will find his/her name in the file named **'RegisteredUser.json'** .After this, the quiz will start. User has to select categories that which type of quiz he/she want to take. The question of quiz are in the file named 'QuestionAnswer.json'

```python
# Function that display main menu of our Quiz App
def QuizAppMenu():
    print("--------------------------------------------------------   ")
    print("         **** Welcome to our Quiz App ****             ")
    print("--------------------------------------------------------\n")
    print("=> Press 1 to Login for the Quiz.")
    print("=> Press 2 to Sign up for the Quiz.")
    print("=> Press 3 to exit.\n")
```

## Sign up Function to get register to take quiz
This function is for sign up for Quiz App, if the user is not already registered. For signing up for Quiz App, the user has to enter his /her name, email address, account password and tell whether you are a student or not. This information will be stored in the 'RegisteredUser.json' file and in this file total quizzes and the last score will also be added.

###### Signup(3)

```python
# Function to sign up for the quiz app
def SignUp(UsersList):

    print("--------------------------------------------")
    print("             Sign Up - Quiz App             ")
    print("--------------------------------------------\n")

    UsersList = readFromRegisteredUserFile(UsersList)

    print("   **** Register for the Quiz **** \n")

    while True:
        UserName = input("=> Enter your Name: ")
        isStudent = input("=> Are you a student? ")
        emailAddress = input("=> Enter your Email Address: ")
        userPassword = input("=> Enter you Password: ")
        totalQuizTaken = 0
        lastScore = 0
        if(checkUserExist(UsersList, emailAddress)):
            print("\nAccount already exist! Try again...\n")
        else:
            break

    User = {
        "UserName": UserName,
        "EmailAddress": emailAddress,
        "userPassword": userPassword,
        "isStudent": isStudent,
        "totalQuizTaken": totalQuizTaken,
        "lastScore": lastScore
    }

    UsersList.append(User)
    UsersDict = {"User": UsersList}
    writeInRegisteredUserFile(UsersDict)
    
    print("\n************ CONGRATULATIONS! YOU ACCOUNT HAS BEEN CREATED. ************\n")
    Login(UsersList)
```

### Utility Function for sign up to validate that the user doesn't already exist

###### SignUp(1)

```python
# Function to check if the user, that is signing up, already exists in our system
def checkUserExist(UsersList, newUserEmail):
    for user in UsersList:
        if user["EmailAddress"] == newUserEmail:
            return True
    return False
```

## Login Function for registered users
If the user is already registered, then he/she will login instead of sign up. In this function, when user tries to login, he/she has to enter his /her email and password from which he/she has sign up. When the user enters his /her name, this function will check all the email and password that are stored in file named 'RegisteredUser.json' until the entered name is not found. If the name not found, then uit means the user has entered the wrong name.

###### Signup(2)

```python
# Function to login for the quiz app
def Login(UsersList):
    print("----------------------------------------------------")
    print("                Login in - Quiz App                 ")
    print("----------------------------------------------------\n")

    UserEmail = input("Enter your Email Address: ")
    UserPassword = input("Enter your Password: ")

    UsersList = readFromRegisteredUserFile(UsersList)

    for checkUser in UsersList:
        if checkUser["EmailAddress"] == UserEmail and checkUser["userPassword"] == UserPassword:
            print(
                "\n**************** CONGRATULATIONS! YOU HAVE LOGGED IN. ****************\n")
            START_QUIZ_FOR_USER(UsersList, UserEmail)
            return 0

    else:
        print("Invalid Credentials.\n")
        print("_________________________________________________________________")
        print("| Press 1 to Sign Up          | Press 2 to login again         |    Press 3 to exit           |")
        print("_________________________________________________________________\n")

        while True:

            wantToRegister = input("Input your Desired Option: ")

            if (wantToRegister == '1'):
                SignUp(UsersList)

            elif (wantToRegister == '2'):
                Login(UsersList)
            
            elif (wantToRegister == '3'):
                break

            else:
                print("Invalid Option: Try Again...\n")
```

## Sub-Menu of Quiz App to select Quiz Category
In this function, you just have to choose the category of quiz.

```python
# Function that display the menu to select Quiz Category 
def QuizMenu():
    print("\n----------------------------------------------------")
    print("                 === Quiz Menu ===                    ")
    print("----------------------------------------------------\n")

    print("Choose a quiz category:")
    print("1. Programming Quiz - Python")
    print("2. General Knowledge Quiz")
    print("3. English Language Quiz")
    print("4. Exit")

    print("\n----------------------------------------------------")
```

## Function to start the quiz of registered user
In this function, we'll implement the Quiz App. We'll select the quiz category in it and then start the quiz. If the user enters the invalid option, a message displays that shows that the user has selected th invalid option and again asks the user to enter the desired option to take the quiz.

###### Start_Quiz_For_User(3)

```python
# Function that implement the quiz app, it will display quiz categories and start quiz according to the category selected by user
def START_QUIZ_FOR_USER(UserList, UserEmail):

    QuizMenu()

    choice = input(
        "\nPlease enter the type of quiz you'd like to take (1-4): ")
    print("\n")

    questionsDict = readFromQuestionAnswerFile()

    TestScore = 0

    while choice != '4':
        if (choice == '1'):
            print("----------------------------------------------------")
            print("            Programming Quiz - Python               ")
            print("----------------------------------------------------\n")
            TestScore = startQuiz(questionsDict["Programming Quiz - Python"])
            updateTheFile(UserList, UserEmail, TestScore)

        elif (choice == '2'):
            print("----------------------------------------------------")
            print("               General Knowledge Quiz               ")
            print("----------------------------------------------------\n")
            TestScore = startQuiz(questionsDict["General Knowledge"])
            updateTheFile(UserList, UserEmail, TestScore)

        elif (choice == '3'):
            print("----------------------------------------------------")
            print("               English Language Quiz                ")
            print("----------------------------------------------------\n")
            TestScore = startQuiz(questionsDict["English Language"])
            updateTheFile(UserList, UserEmail, TestScore)

        else:
            print("No such option exist.")

        choice = input(
            "\nWould you like to take quiz again? (1-4): ")
        print("\n")
```

## Function to take quiz as per category selected by user
This function will take the quiz from user and tell him/her whether he has choose the correct option or not. In the last, this function will display the total score gained by him/her in this quiz.

###### Start_Quiz_For_User(1)

```python
# function that take quiz from the user as per the category selected by the user
def startQuiz(quizCategory):
    TestScore = 0
    i = 1
    
    for quizQuestions in quizCategory:
        print(str(i) + "). " + quizQuestions["Question"])
        print("    A). " + quizQuestions["A)"],)
        print("    B). " + quizQuestions["B)"])
        print("    C). " + quizQuestions["C)"])
        print("    D). " + quizQuestions["D)"], "\n")

        SelectedAnswer = input("Enter Your Answer: ")

        correctAnswer = quizQuestions["CorrectAnswer"]

        if (SelectedAnswer == quizQuestions["CorrectAnswer"].lower() or SelectedAnswer == quizQuestions["CorrectAnswer"].upper()):
            print("*** CORRECT ANSWER ***\n")
            TestScore += 1
            
        else:
            print(f"INCORRECT ANSWER: Correct option is {correctAnswer}\n")
        i += 1

    print("----------------------------------------------------")
    print("                    Quiz Result                     ")
    print("----------------------------------------------------")
    print(f"You have scored {TestScore} out of 5 marks. ")
    print("----------------------------------------------------\n")

    return TestScore
```

### Function that updates the "RegisteredUser.json" File everytime a user has taken quiz
In this function, we'll update the lastscore and the total quizzes taken in file named 'RegisteredUser.json'.

###### Start_Quiz_For_User(2)

```python
# Function that update user test score and no of quiz taken data in "RegisteredUser.json" file
def updateTheFile(UsersList, UserEmail, TestScore):
    for user in UsersList:
        if user["EmailAddress"] == UserEmail:
            user["lastScore"] = TestScore
            user["totalQuizTaken"] += 1
    UsersDict = {"User": UsersList}
    writeInRegisteredUserFile(UsersDict)
```

After selecting the category, we'll given questions from that category and we've to entered the correct option. If we've not enter the correct option, the message of incorrect answer is displayed. And, also tell us the correct option.

After successful completion of quiz, your marks will be displayed and asks whether you want to take the quiz again or not.If you want to take the quiz again, choose the category otherwise exit the program.
